```python
datasets = pd.read_csv('data_summary.csv', sep=',', dtype={'dataset': str})
datasets['dataset']
```

```python
fig = px.box(datasets, x="dataset", y="y", color='dataset')
fig.show()
```

```python
fig = px.scatter(datasets, x='dataset', y='y', title='Scatterplot comparision of the datasets')
fig.show()
```

Boxplot is not useful for h_lines, high_lines and dots, compared with other, because they form a different type of distributuon.
